# 02 Exploratory Data Analysis
Detailed analysis of student performance, placement testing, and enrollment holds.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

students = pd.read_csv('../data/synthetic/students.csv')
holds = pd.read_csv('../data/synthetic/enrollment_holds.csv')

# Status Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=students, x='student_status', palette='viridis')
plt.title('Status Distribution')
plt.show()
```

## Hold Resolution Analysis

```python
avg_days = holds.groupby('hold_type')['days_to_resolve'].mean()
avg_days.plot(kind='bar', color='salmon')
plt.title('Avg Days to Resolve by Hold Type')
plt.show()
```
