# 01 Data Loading and Cleaning
This notebook covers the loading and initial cleaning of the UCI student success dataset and the generated synthetic datasets.

**Note:** Before running this notebook, make sure you have downloaded the public dataset **'Predict Students’ Dropout and Academic Success'** from the UCI Machine Learning Repository and placed it at `data/raw/student_success_uci.csv`.

```python
import pandas as pd
import os

# Define paths
RAW_UCI = '../data/raw/student_success_uci.csv'
SYNTH_STUDENTS = '../data/synthetic/students.csv'

# Check if UCI Data exists
if not os.path.exists(RAW_UCI):
    raise FileNotFoundError(
        f"Raw UCI dataset not found at: {RAW_UCI}\n"
        "Please download the 'Predict Students' Dropout and Academic Success' dataset manually\n"
        "from the UCI Machine Learning Repository and place it at that location.\n"
        "Refer to the project README.md or data/README.md for detailed instructions."
    )

# Detect delimiter (the real UCI dataset is typically semicolon-delimited)
try:
    with open(RAW_UCI, 'r', encoding='utf-8') as f:
        first_line = f.readline()
    sep = ';' if ';' in first_line else ','
except Exception:
    sep = ','

# Load UCI Data
df_uci = pd.read_csv(RAW_UCI, sep=sep)
print(f"UCI Data Shape: {df_uci.shape}")

# Cleaning UCI Data
df_uci.columns = df_uci.columns.str.strip().str.lower().str.replace(' ', '_')
df_uci = df_uci.rename(columns={'target': 'student_status'})
df_uci.head()
```
