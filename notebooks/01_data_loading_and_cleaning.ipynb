# 01 Data Loading and Cleaning
This notebook covers the loading and initial cleaning of the UCI student success dataset and the generated synthetic datasets.

```python
import pandas as pd
import os

# Define paths
RAW_UCI = '../data/raw/student_success_uci.csv'
SYNTH_STUDENTS = '../data/synthetic/students.csv'

# Load UCI Data
df_uci = pd.read_csv(RAW_UCI)
print(f"UCI Data Shape: {df_uci.shape}")

# Cleaning UCI Data
df_uci.columns = df_uci.columns.str.lower().str.replace(' ', '_')
df_uci = df_uci.rename(columns={'target': 'student_status'})
df_uci.head()
```
